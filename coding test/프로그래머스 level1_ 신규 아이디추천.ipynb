```python
def solution(new_id):
    answer = ''
    # 1단계 대문자 -> 소문자
    s1 = new_id.lower()
    # 2단계 특정문자외 제거
    for value in s1:
        if value.islower() or value.isdigit() or value in ['-','_','.'] :
            answer += value
    # 3단계 .2번이상 . 하나로 바꾸기
    while '..' in answer :
        answer = answer.replace('..','.')
    # 4단계 . 처음이나 끝이면 제거
    if answer[0] =='.' :
        if len(answer) >= 2:
            answer = answer[1:]
        else :
            answer = '.'
    if answer[-1] == '.':
        answer = answer[:-1]
    # 5단계 new_id가 빈 문자열이면 a 대입
    if answer == '':
        answer ='a'
    # 6단계 아이디 길이가 16자 이상이면 그 이후로 제거
    if len(answer) >= 16 : 
        answer = answer[:15]
        if answer[-1] == '.':
            answer= answer[:-1]
    # 7단계 new_id 길이가 2자 이하면, new_id의 마지막 문자를 new_id의 길이가 3이 될때까지 반복
    while len(answer) <= 2 :
        answer += answer[-1]
                
    return answer
```
