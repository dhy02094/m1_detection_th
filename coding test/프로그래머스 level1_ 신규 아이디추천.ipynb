```python
def solution(new_id):
    answer = ''
    # 1  -> 
    s1 = new_id.lower()
    # 2  
    for value in s1:
        if value.islower() or value.isdigit() or value in ['-','_','.'] :
            answer += value
    # 3 .2 .  
    while '..' in answer :
        answer = answer.replace('..','.')
    # 4 .   
    if answer[0] =='.' :
        if len(answer) >= 2:
            answer = answer[1:]
        else :
            answer = '.'
    if answer[-1] == '.':
        answer = answer[:-1]
    # 5 new_id   a 
    if answer == '':
        answer ='a'
    # 6   16    
    if len(answer) >= 16 : 
        answer = answer[:15]
        if answer[-1] == '.':
            answer= answer[:-1]
    # 7 new_id  2 , new_id   new_id  3  
    while len(answer) <= 2 :
        answer += answer[-1]
                
    return answer
```
